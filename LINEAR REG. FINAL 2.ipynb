```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class LinearRegression:
    def __init__(self, lr=0.01, iters=1000):
        self.lr, self.iters = lr, iters
    
    def fit(self, X, y):
        bias = 0
        weights = 0
        n = X.shape[0]
        cost_list=[]
        for i in range(self.iters):
            d_bias = -2 * np.sum(y - weights*X + bias) / n
            d_weights = -2 * np.sum(X*(y - (weights*X + bias))) / n
            bias = bias + (self.lr * d_bias)
            weights= weights - (self.lr * d_weights)
            y_pred=weights*X + bias
            cost=(-1/(2*n))*np.sum(y_pred-y)
            cost_list.append(cost)
        self.weights=weights
        self.bias=bias
        self.costs=cost_list
        
    def predict(self, X):
        return self.weights*X + self.bias
    def get_costs(self):
        return self.costs
```

```python
np.random.seed(10)
X = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.5)
y = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.25)
```

```python
plt.scatter(X,y)
```

```python
clf = LinearRegression()
clf.fit(X, y)
```

```python
plt.scatter(X, y)
plt.plot(X, clf.predict(X),c='r')
plt.title("Gradient Descent Linear Regression")
```

```python
costs=clf.get_costs()[:150]
iters=list(range(150))
```
