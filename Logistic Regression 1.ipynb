```python
n=int(0.8*len(df2))

X_train=df2[:n]
X_train=np.array(X_train.drop(columns='TenYearCHD'))
X_test=df2[n:]
X_test=np.array(X_test.drop(columns='TenYearCHD'))

y_train=np.array(df2['TenYearCHD'][:n])
y_train=y_train.reshape(-1,1)
y_test=np.array(df2['TenYearCHD'][n:])
y_test=y_test.reshape(-1,1)
```

```python
class LogisticRegression:
    def __init__(self,lr=0.001,n_iters=1000):
        self.lr=lr
        self.n_iters=n_iters
        self.weights=None
        self.bias=None
        
    def logistic(self,x):
        return 1/(1+np.exp(-x))
    
    def fit(self,X,y):
        self.n_samples,self.n_features=X.shape
        self.weights=np.zeros(self.n_features).reshape(1,-1)
        
        self.bias=0
        for i in range(self.n_iters):
            y_pred=self.logistic(np.dot(self.weights,X.T) + self.bias)
            
            #cost=(-1/n_samples)*(np.dot(y.T,np.log(y_pred)))+np.dot((1-y.T),np.log(1-y_pred))
            self.weights=self.weights-self.lr*np.dot((y_pred-y.T),X)
            #pdb.set_trace()
            self.bias=self.bias-self.lr*np.sum((y_pred-y.T))
        print("Weights:",self.weights)
        print("Bias:", self.bias)

    
    def predict(self,X):
        y_pred=self.logistic(np.dot(X,self.weights.T) + self.bias)
        #print(y_pred)
        l=[]
        l=[1 if i>=0.5 else 0 for i in y_pred]
        return l
    
    def check(self):
        print(self.weights)
        print(self.bias)
        print(self.n_samples)
        print(self.n_features)
```

```python
regressor=LogisticRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
regressor.predict(X_test)
```
